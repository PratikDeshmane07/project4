```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# Replace 'your_file_path.csv' with the path to your dataset
heart_data = pd.read_csv('cleaned data/kaggle data/processed_kaggleheart.csv')
heart_data
```

```python
# Sample mapping (modify according to your specific needs)
heart_data['Sex'] = heart_data['Sex'].map({'M': 'Male', 'F': 'Female'})
heart_data['ExerciseAngina'] = heart_data['ExerciseAngina'].map({'Y': 'Yes', 'N': 'No'})
heart_data['ChestPainType'] = heart_data['ChestPainType'].map({'ATA': 'Atypical Angina', 
                                                               'NAP': 'Non-Anginal Pain', 
                                                               'ASY': 'Asymptomatic', 
                                                               'TA': 'Typical Angina'})
heart_data['RestingECG'] = heart_data['RestingECG'].map({'Normal': 'Normal', 
                                                         'ST': 'ST-T Wave Abnormality', 
                                                         'LVH': 'Left Ventricular Hypertrophy'})
heart_data['ST_Slope'] = heart_data['ST_Slope'].map({'Up': 'Upsloping', 
                                                     'Flat': 'Flat', 
                                                     'Down': 'Downsloping'})

# Display the modified DataFrame to check the changes
heart_data
```

```python
encoder = OneHotEncoder(sparse_output=False)
categorical_columns = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']
encoded_data = pd.DataFrame(encoder.fit_transform(heart_data[categorical_columns]))

# Update column names to reflect encoded features
encoded_data.columns = encoder.get_feature_names_out(categorical_columns)

# Drop original categorical columns
heart_data_encoded = heart_data.drop(categorical_columns, axis=1)

# Concatenate encoded columns
onehotencodingheartdata = pd.concat([heart_data_encoded, encoded_data], axis=1)

onehotencodingheartdata

onehotencodingheartdata.to_csv('cleaned data/kaggle data/encoded_kaggleheart.csv', index=False)
```
